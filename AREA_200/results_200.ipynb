## AREA : (200 * 200)

### 500 NODES

#### Voronoi-Q-learning Positioninig for Initial Sensor Deployment

```python
sim1 = Simulation(
    area_size=(200, 200),
    n_nodes=500,
    rounds=6000,
    init_energy=0.5,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="adaptive",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    localization_mode="DRL"
)
metrics = sim1.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim1.save_results("wsn_500_DRL_results.csv")
sim1.plot()
```

#### Random Positioning for Initial Sensor Deployment

```python
sim2 = Simulation(
    area_size=(200, 200),
    n_nodes=500,
    rounds=6000,
    init_energy=0.5,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="adaptive",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    localization_mode="random"
)
metrics = sim2.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim2.save_results("wsn_500_random_results.csv")
sim2.plot()
```
