## Random deployment, 200 nodes, 1J initial energy

### optimizer based

```python
sim1 = Simulation(
    area_size=(100, 100),
    n_nodes=200,
    rounds=60000,
    init_energy=1,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="eeosp",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    head_selection_strategy="optimizer",
    localization_mode="random"
)
metrics = sim1.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim1.save_results("wsn_optimizer_results.csv")
sim1.plot()
```

### adaptive based

```python
sim3 = Simulation(
    area_size=(100, 100),
    n_nodes=200,
    rounds=60000,
    init_energy=1,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="eeosp",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    head_selection_strategy="adaptive",
    localization_mode="random"
)
metrics = sim3.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim3.save_results("wsn_adaptive_results.csv")
sim3.plot()
```
