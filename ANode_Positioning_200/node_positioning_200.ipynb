## 200 Nodes

### AREA -> 200x200

#### Voronoi-Q-learning Positioninig for Initial Sensor Deployment

```python
sim = Simulation(
    area_size=(200, 200),
    n_nodes=200,
    rounds=60000,
    init_energy=1,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="adaptive",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    localization_mode="DRL"
)
metrics = sim.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim.save_results("wsn_200_200_DRL_results.csv")
sim.plot()
```

#### Random Positioning for Initial Sensor Deployment

```python
sim1 = Simulation(
    area_size=(200, 200),
    n_nodes=200,
    rounds=60000,
    init_energy=1,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="adaptive",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    localization_mode="random"
)
metrics = sim1.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim1.save_results("wsn_200_200_random_results.csv")
sim1.plot()
```

#### Comparative Plot

```python
plot_comparison([sim, sim1], ["Q-learning", "random"])
```
