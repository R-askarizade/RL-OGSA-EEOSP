## AREA : (100 * 100)

### 100 NODES - Initial Energy : 0.5J

#### Voronoi-Q-learning Positioninig for Initial Sensor Deployment

```python
sim = Simulation(
    area_size=(100, 100),
    n_nodes=100,
    rounds=60000,
    init_energy=0.5,
    comm_range=50.0,
    recluster_period=50,
    sink_mode="adaptive",
    sink_visit_period=5,
    routing_mode="multi-hop",
    seed=42,
    localization_mode="DRL"
)
metrics = sim.run()
print("\n=== Simulation Summary ===")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

sim.save_results("wsn_100_DRL_results.csv")
sim.plot()
```

#### Random Positioning for Initial Sensor Deployment

#### Comparative Plot

```python
plot_comparison([sim, sim1], ["Q-learning", "random"])
```
